````python
# Find the name of the class with the top score when mean-aggregated across frames.
def class_names_from_csv(class_map_csv_text):
  """Returns list of class names corresponding to score vector."""
  class_names = []
  with tf.io.gfile.GFile(class_map_csv_text) as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
      class_names.append(row['display_name'])

  return class_names

class_map_path = model.class_map_path().numpy()
class_names = class_names_from_csv(class_map_path)
````

````python
def ensure_sample_rate(original_sample_rate, waveform,
                       desired_sample_rate=16000):
  """Resample waveform if required."""
  if original_sample_rate != desired_sample_rate:
    desired_length = int(round(float(len(waveform)) /
                               original_sample_rate * desired_sample_rate))
    waveform = scipy.signal.resample(waveform, desired_length)
  return desired_sample_rate, waveform
````

````python
def process_audio_file(file_path, class_names):
    sample_rate, wav_data = wavfile.read(file_path)
    sample_rate, wav_data = ensure_sample_rate(sample_rate, wav_data)
    waveform = wav_data / tf.int16.max
    scores, embeddings, spectrogram = model(waveform)

    scores = scores.numpy()
    time_stamps = []
    threshold = 0.4
    frame_duration = 0.96  # Each frame is 0.96 seconds long in YAMNet

    # Iterate over the frames
    for i, score in enumerate(scores):
        top_index = np.argmax(score)
        top_score = score[top_index]
        if top_score > threshold:
            class_name = class_names[top_index]
            start_time = i * frame_duration
            end_time = start_time + frame_duration
            time_stamps.append((class_name, start_time, end_time))

    # Aggregate contiguous time intervals
    aggregated_intervals = defaultdict(list)
    for class_name, start_time, end_time in time_stamps:
        if aggregated_intervals[class_name] and aggregated_intervals[class_name][-1][1] == start_time:
            aggregated_intervals[class_name][-1] = (aggregated_intervals[class_name][-1][0], end_time)
        else:
            aggregated_intervals[class_name].append((start_time, end_time))

    # Calculate total duration for each class
    class_durations = {class_name: sum(end - start for start, end in intervals) for class_name, intervals in aggregated_intervals.items()}

    # Sort classes by total duration
    sorted_classes = sorted(class_durations.items(), key=lambda item: item[1], reverse=True)

    return aggregated_intervals, sorted_classes
````

````python



def detect_classes_with_durations(wav_file_name):
    sample_rate, wav_data = wavfile.read(wav_file_name, 'rb')
    sample_rate, wav_data = ensure_sample_rate(sample_rate, wav_data)
    waveform = wav_data / tf.int16.max
    scores, embeddings, spectrogram = model(waveform)
    
    # Calculate mean scores for each class
    mean_scores = np.mean(scores, axis=0)
    
    # Initialize tracking variables
    class_durations = defaultdict(float)
    class_start_times = defaultdict(lambda: float('inf'))
    class_end_times = defaultdict(lambda: float('-inf'))
    
    # Assume window size and step size (should match model's parameters)
    window_size = 0.976  # Example: 0.976 seconds per frame
    step_size = 0.976    # Example: 0.976 seconds per step
    
    # Process scores to determine class presence
    for idx, score in enumerate(mean_scores):
        class_name = class_names[idx]
        time = idx * step_size
        if score > 0.2:  # Threshold for class presence
            class_durations[class_name] += step_size
            class_start_times[class_name] = min(class_start_times[class_name], time)
            class_end_times[class_name] = max(class_end_times[class_name], time + step_size)
    
    # Calculate total duration for each class
    total_durations = {class_name: end - start for class_name, start, end in zip(class_durations.keys(), class_start_times.values(), class_end_times.values())}

    # Aggregate scores by class
    score_aggregate = defaultdict(float)
    for idx, score in enumerate(mean_scores):
        class_name = class_names[idx]
        score_aggregate[class_name] += score

    # Sort classes by total score in descending order
    sorted_classes = sorted(score_aggregate.items(), key=lambda item: item[1], reverse=True)

    # Get the top 3 classes
    top_classes = sorted_classes[:3]

    # Print the top 3 classes with their scores and durations
    for i, (class_name, score) in enumerate(top_classes):
        duration = total_durations.get(class_name, 0)
        print(f"Top {i + 1} class: {class_name}, Score: {score:.4f}, Duration: {duration:.2f}s")

# Example usage
detect_classes_with_durations("/home/focus/Bureau/Audio_Node/recordings/test1")
````

````python
import pyaudio
import wave
import os
import time

# Constants
CHUNK = 4096
FORMAT = pyaudio.paInt16
CHANNELS = 1
RATE = 16000
DURATION = 10  # Duration of each recording in seconds
OUTPUT_FOLDER = "/home/focus/Bureau"  # Folder where recordings will be saved

# Create the output folder if it does not exist
if not os.path.exists(OUTPUT_FOLDER):
    os.makedirs(OUTPUT_FOLDER)

def record_audio(file_path):
    p = pyaudio.PyAudio()
    stream = p.open(format=FORMAT,
                    channels=CHANNELS,
                    rate=RATE,
                    input=True,
                    frames_per_buffer=CHUNK)

    frames = []

    print("Recording...")
    for _ in range(int(RATE / CHUNK * DURATION)):
        data = stream.read(CHUNK, exception_on_overflow=False)
        frames.append(data)

    print("Recording finished.")
    stream.stop_stream()
    stream.close()
    p.terminate()

    with wave.open(file_path, 'wb') as wf:
        wf.setnchannels(CHANNELS)
        wf.setsampwidth(p.get_sample_size(FORMAT))
        wf.setframerate(RATE)
        wf.writeframes(b''.join(frames))

def main():
    index = 1
    while True:
            file_path = os.path.join(OUTPUT_FOLDER, f"{index}.wav")
            record_audio(file_path)
            index += 1
            time.sleep(1)  # Optional delay before starting the next recording
        

if __name__ == "__main__":
    main()
````
